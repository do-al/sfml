```python
# Составим массив из названий колонок с целочисленными типами (в нашем случае это будут все колонки)
all_columns = list(df.columns.values) # имена всех колонок
int_columns = all_columns
for i in range(len(int_columns)):
    if not df[int_columns[i]].dtype == 'int64':
        int_columns.remove(int_columns[i])
# имена целочисленных колонок
print (int_columns)
```

```python
# Статистическая функция, вычисляющая для заданного числового столбца заданной таблицы количество элементов, 
# находящихся рядом с НИЖНЕЙ границей диапазона значений массива (ближе, чем на N% от нижней границы)
def near_low_bound (any_table, any_column, percent):
    min_value = any_table[any_column].min()
    max_value = any_table[any_column].max()
    delta = (max_value - min_value)*percent/100
    a = any_table[any_table[any_column] < min_value + delta][any_column].count()
    #print(min_value)
    #print(max_value)
    #print(delta)
    return a
```

```python
near_low_bound(df, 'ID', 10)
```

```python
# Статистическая функция, вычисляющая для заданного числового столбца заданной таблицы количество элементов, 
# находящихся рядом с ВЕРХНЕЙ границей диапазона значений массива (ближе, чем на N% от верхней границы)
def near_up_bound (any_table, any_column, percent):
    min_value = any_table[any_column].min()
    max_value = any_table[any_column].max()
    delta = (max_value - min_value)*percent/100
    a = any_table[any_table[any_column] > max_value - delta][any_column].count()
    return a
```

```python
near_up_bound(df, 'ID', 10)
```

```python
# Заполняем 2 массива со значениями функций near_low_bound и near_up_bound
low_values = []
up_values = []
i = 0
while (i < len(int_columns)):
    low_values.append(near_low_bound(df, int_columns[i], 10))
    up_values.append(near_up_bound(df, int_columns[i], 10))
    i = i + 1
```

```python
# Создаем таблицу с именами целочисленных колонок и значениями статистических функций near_low_bound и near_up_bound
df2 = pd.DataFrame({
    'LOW_VALUE_COUNT':low_values,
    'UP_VALUE_COUNT':up_values
}, index=int_columns)
```

```python
# (6) Создать новый строковый столбец в data frame-е, который:
# принимает значение A, если значение LIMIT_BAL <=10000
# принимает значение B, если значение LIMIT_BAL <=100000 и >10000
# принимает значение C, если значение LIMIT_BAL <=200000 и >100000
# принимает значение D, если значение LIMIT_BAL <=400000 и >200000
# принимает значение E, если значение LIMIT_BAL <=700000 и >400000
# принимает значение F, если значение LIMIT_BAL >700000

#TODO
```

```python
def ball_values(n):
    if n<=10000:
        s='A'
    elif n<=100000:
        s='B'
    elif n<=200000:
        s='C'
    elif n<=400000:
        s='D'
    elif n<=700000:
        s='E'
    else:
        s='F'
    return s   
```

```python
df8 = df # Копия исходной таблицы
```

```python
df8['LIMIT_BALL_INDICATOR'] = df8['LIMIT_BAL'].map(ball_values)
```
